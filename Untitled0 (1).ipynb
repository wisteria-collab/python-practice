```python
first = input( "最初の数字を入力してください")
next = input("次の数字を入力してください")
result_puls= int(first) + int(next)
print(result_puls)
result_minus= int(first) - int(next)
print(result_minus)
result_times= int(first) * int(next)
print(result_times)
result_divide= int(first) / int(next)
print(result_divide)
```
